```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"F:\Self Learning\Code_basics_projects\Wine Quality Dataset\winequalityN.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.fillna(0)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df.type = le.fit_transform(df.type)
```

```python
df.head()
```

```python
x = df.drop("quality", axis=1)
y = df["quality"]
```

```python
from sklearn.preprocessing import StandardScaler
Standar_x = StandardScaler()
x = Standar_x.fit_transform(x)
```

```python
print(x), print(y)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, roc_auc_score
```

```python
x_train, x_test, y_train,  y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
KNN = KNeighborsClassifier()
```

```python
parameters = {
    'n_neighbors': [3, 5, 7, 9], 
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}
```

```python
classifier = GridSearchCV(KNN, param_grid=parameters, scoring = "accuracy", cv=5)
classifier.fit(x_train, y_train)
```

```python
classifier.get_params()
```

```python
y_hat = classifier.predict(x_test)
```

```python
classifier.score(x_test, y_test)
```

```python
classifier.best_params_
```

```python
classifier.best_score_
```

```python
score = accuracy_score(y_test, y_hat)
print(score)
```

```python
classifier.score(x_train, y_train)
```

```python
report = classification_report(y_test, y_hat)
print(report)
```

```python
pre = precision_score(y_test, y_hat, average="micro")
print(pre)
```

```python
re = recall_score(y_test, y_hat, average="micro")
print(re)
```
