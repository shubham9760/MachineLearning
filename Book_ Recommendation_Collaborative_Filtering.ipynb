# ***Collaborative filtering***

```python
ratings_with_name.head(2)
```

***Selecting only those users who have given ratings more than 200***

```python
x = ratings_with_name.groupby("User-ID")["Book-Rating"].count()>200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(padhe_likhe_users)]
filtered_rating
```

***Here we are selecting those books whose rating count is greater than 50***

```python
y = filtered_rating.groupby("Book-Title")["Book-Rating"].count()>=50
famous_books = y[y].index
```

```python
filtered_rating = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
filtered_rating.head(100)
```

```python
pt = filtered_rating.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
```

```python
pt.shape
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score= cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    simmilar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    data=[]
    for i in simmilar_items:
        item=[]
        temp_df = books[books["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))

        data.append(item)
    return data
```

```python
recommend("1984")
```

```python
pickle.dump(pt, open("pt.pkl", "wb"))
pickle.dump(books, open("books.pkl", "wb"))
pickle.dump(similarity_score, open("similarity_score.pkl", "wb"))
```

# ***Model Building***

***Here we are going to use item-item collaborative filtering***

```python
from sklearn.neighbors import NearestNeighbors

df_transposed = pt

def recommendation(book_name):
    index = np.where(pt.index==book_name)[0][0]
    k = 6
    model = NearestNeighbors(n_neighbors=k, metric='cosine')
    model.fit(df_transposed)

    distances, indices = model.kneighbors(df_transposed.iloc[[index]])

    recommended_books = []
    for neighbor_index in indices.flatten():
        if neighbor_index != index:
            recommended_books.append(pt.index[neighbor_index])

    return recommended_books

```

```python
book_name = input("Enter the name of the book")
recommended_books = recommendation(book_name)
print("Recommended books similar to", book_name, "are :\n")
for i in recommended_books:
    print(i)
```
